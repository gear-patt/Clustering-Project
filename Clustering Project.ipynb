```python
import os
import pandas as pd
os.chdir('data')
data = pd.read_csv('Human_Activity_Recognition_Using_Smartphones_Data.csv')
data.head()
```

```python
original_data = data.copy()
```

```python
data = data.drop('Activity', axis=1)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
data = s.fit_transform(data)
```

```python
data
```

```python
from sklearn.decomposition import PCA
import numpy as np
pca_list = list()
feature_weight_list = list()

# Fit a range of PCA models

for n in [25, 50, 100, 200, 300, 400, 500]:
    
    # Create and fit the model
    PCAmod = PCA(n_components=n)
    PCAmod.fit(data)
    
    # Store the model and variance
    pca_list.append(pd.Series({'n':n, 'model':PCAmod,
                               'var': PCAmod.explained_variance_ratio_.sum()}))
    
    
pca_df = pd.concat(pca_list, axis=1).T.set_index('n')
pca_df
```

You can see that using PCA reducing the number of features from 561 features to 200 or 100 doesn't change the variance much. So, we can choose to use just 100 features to avoid curse of dimensionality.

```python
data.shape
```

```python
pca = PCA(n_components=100)
pca.fit(data)
data = pca.transform(data)
data
```

```python
data.shape
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
inertia = []
list_num_clusters = list(range(1,11))
for num_clusters in list_num_clusters:
    km = KMeans(n_clusters=num_clusters)
    km.fit(data)
    inertia.append(km.inertia_)
    
plt.plot(list_num_clusters,inertia)
plt.scatter(list_num_clusters,inertia)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia');
```

Since number of clusters=2 is the elbow point, we choose n_clusters=2.

```python
km = KMeans(n_clusters=2)
km = km.fit(data)
km1 = km.predict(data)
km1
```

```python
km = KMeans(n_clusters=6)
km = km.fit(data)
km2 = km.predict(data)
km2
```

```python
original_data['km1'] = km1
original_data
```

```python
(original_data[['Activity', 'km1']]
 .groupby(['Activity','km1'])
 .size()
 .to_frame()
 .rename(columns={0:'number'}))
```

```python
original_data['km2'] = km2
(original_data[['Activity', 'km2']]
 .groupby(['Activity','km2'])
 .size()
 .to_frame()
 .rename(columns={0:'number'}))
```

You can see that the model segment the data into 2 groups, which makes sense because if the activity is given, the data is cleary seperated into moving and not moving.

```python
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(n_clusters=2, linkage='ward', compute_full_tree=True)
ag = ag.fit(data)
ag1 = ag.fit_predict(data)
ag1
```

```python
original_data['ag1'] = ag1
(original_data[['Activity', 'ag1']]
 .groupby(['Activity','ag1'])
 .size()
 .to_frame()
 .rename(columns={0:'number'}))
```

```python
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(n_clusters=6, linkage='ward', compute_full_tree=True)
ag = ag.fit(data)
ag2 = ag.fit_predict(data)
ag2
```

```python
original_data['ag2'] = ag2
(original_data[['Activity', 'ag2']]
 .groupby(['Activity','ag2'])
 .size()
 .to_frame()
 .rename(columns={0:'number'}))
```

```python
x_with_agglo = original_data.drop(['Activity', 'km1', 'km2', 'ag2'], axis=1)
x_without_agglo = x_with_agglo.drop('ag1', axis=1)
y = original_data['Activity']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import classification_report
s = StratifiedShuffleSplit(n_splits=1, random_state=0)

def train_models(x,y):
    train_idx, test_idx = next(s.split(x,y))
    x_train = x.iloc[train_idx,:]
    y_train = y.iloc[train_idx]
    x_test = x.iloc[test_idx,:]
    y_test = y.iloc[test_idx]
    lr = LogisticRegression(solver='liblinear').fit(x_train, y_train)
    y_predict = lr.predict(x_test)
    return classification_report(y_test, y_predict)
```
